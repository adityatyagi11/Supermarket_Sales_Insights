# EXPLORATORY DATA ANALYSIS

```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the Dataset

```python
df = pd.read_csv("E:\\TECHSTACK\\supermarket_sales.csv")
```

Here's a brief description of each of the attributes or labels in the dataset:

Invoice ID: A unique identifier for each invoice or transaction.

Branch: The branch or location where the transaction occurred.

City: The city where the branch is located.

Customer Type: Indicates whether the customer is a regular or new customer.

Gender: The gender of the customer.

Product Line: The category or type of product purchased.

Unit Price: The price of a single unit of the product.

Quantity: The number of units of the product purchased.

Tax 5%: The amount of tax (5% of the total cost) applied to the transaction.

Total: The total cost of the transaction, including tax.

Date: The date when the transaction took place.

Time: The time of day when the transaction occurred.

Payment: The payment method used (e.g., credit card, cash).

COGS (Cost of Goods Sold): The direct costs associated with producing or purchasing the products sold.

Gross Margin Percentage: The profit margin percentage for the transaction.

Gross Income: The total profit earned from the transaction.

Rating: Customer satisfaction rating or feedback on the transaction.

```python
#converting the type of date column to datetime
df1["Date"] = pd.to_datetime(df["Date"])
```

# Visualization

```python
#Sales over time
daily_sales = df1.groupby("Date")["Total"].sum()


plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title("DAILY SALES OVER TIME")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Total sales by product line
product_sales = df1.groupby("Product line")["Total"].sum().sort_values()

plt.figure(figsize=(10, 6))
product_sales.plot(kind="barh", color="skyblue")
plt.title("TOTAL SALES BY PRODUCT LINE")
plt.xlabel("TOTAL SALES")
plt.ylabel("PRODUCT LINE")
plt.show()
```

```python
#product line vs gross income
plt.figure(figsize=(13,6))
sns.barplot(x = df1["Product line"],y= df1["gross income"])
plt.title("PRODUCT LINE VS GROSS INCOME")
plt.show()
```

```python
#Product line vs Gender
df1.groupby(["Gender","Product line"]).count()["Invoice ID"]
sales_gender = df1.groupby(["Gender","Product line"]).count()["Invoice ID"]
sales_gender.unstack(level = 0).plot(kind = 'bar') 
plt.legend(loc = "upper right")  
plt.grid(axis = "y")
plt.title("PRODUCT LINE VS GENDER")
plt.show()
```

```python
#Sales by gender and customer type
plt.figure(figsize=(8, 6))
sns.barplot(x="Gender", y="Total", hue="Customer type", data=df1)
plt.title("SALES BY GENDER AND CUSTOMER TYPE")
plt.show()
```

```python
# Payment method distribution
payment_method_counts = df1["Payment"].value_counts()

plt.figure(figsize=(8, 4))
sns.barplot(x=payment_method_counts.index, y=payment_method_counts.values)
plt.title("PAYMENT METHOD DISTRIBUTION")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
# Total sales by branch
branch_sales = df1.groupby("Branch")["Total"].sum().sort_values()


plt.figure(figsize=(8, 6))
branch_sales.plot(kind="bar")
plt.title("TOTAL SALES BY BRANCH")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Genderwise distribution of data
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df1, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("DISTRIBUTION OF GENDER")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#product line vs rating
plt.figure(figsize=(12,6)) 
sns.barplot(y = df1["Product line"],x= df1["Rating"])
plt.title("PRODUCT LINE VS RATING")
plt.show()
```

# Conclusion and Insights

The exploratory data analysis of the supermarket sales dataset has provided valuable insights into various aspects of sales performance and customer behavior. The key findings from the analysis are summarized below:

1- Sales Trends and Peak Periods: The identification of peak sales periods indicates that the supermarket experiences fluctuations in daily sales. These fluctuations are likely influenced by factors such as weekdays, holidays, and promotional events. Understanding these patterns can help the supermarket optimize staffing, inventory management, and promotional strategies to align with peak periods.

2- Product Line Performance: The "Food and Beverages" product line has emerged as a major revenue driver, significantly outperforming other product lines. This suggests a strong customer preference for these items. To capitalize on this, the supermarket should prioritize marketing efforts and stock management for these products, potentially increasing overall sales and customer satisfaction.

3- The gross income data reveals that the "Home and Lifestyle" product line generates the highest revenue, followed by "Sports and Travel," with "Food and Beverages" in third place. This distribution indicates that while "Food and Beverages" is a major revenue driver, "Home and Lifestyle" products are currently the most profitable.

4- Demographic Influences:

Gender-Based Preferences: Female shoppers predominantly purchase "Fashion Accessories," while male shoppers are more inclined toward "Health and Beauty" products. This gender-based preference highlights the importance of tailoring marketing and inventory strategies to align with these preferences, potentially increasing sales and customer satisfaction.

5- Membership Impact: Membership status also affects sales, with higher contributions from members. Developing targeted campaigns and offers for members could further boost sales.

6- Payment Method Preferences: The high use of "Ewallet" for transactions indicates a strong customer preference for digital payments. Enhancing the availability and functionality of digital payment options can improve the customer experience and streamline the checkout process.

7- Branch Performance: Significant disparities in sales performance among branches were noted, with "Branch C" showing the highest sales. This suggests that branch-specific factors, such as location and demographic profiles, impact performance. Further investigation into these factors could help tailor strategies to improve results across all branches.

8- Gender Distribution: The customer base comprises 501 females and 499 males, indicating a nearly equal gender distribution.

9- Customer Ratings and Satisfaction: The generally high customer ratings reflect positive satisfaction.
